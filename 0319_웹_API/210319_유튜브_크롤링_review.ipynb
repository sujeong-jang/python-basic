```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome() # 브라우저 권한 할당 (자동으로 브라우저 실행)
```

1. 화면에서 키워드를 받아서 해당 키워드로 검색

```python
search_keyword = input("검색어: ")
```

```python
search_keyword
```

```python
url = "https://www.youtube.com/results?search_query=" + search_keyword
```

```python
# 2. url에서 driver 가져오기
```

```python
driver.get(url)
html = driver.page_source # 데이터 불러오기
```

```python
# 3. html에 parser 적용
from bs4 import BeautifulSoup
soup = BeautifulSoup(html)
```
