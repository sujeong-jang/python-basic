```python
import json
import xmltodict # xml -> dict
import pandas as pd
import requests

# params로 접근
my_key = "0C6iX2Ww%2BnB9GVY6k9Et5H9zqv%2FxjEsldVq1Wa9Co3lvcuaRi%2BI3gcoFoi3AckHg6B8AAf1cH0tKH4MXhYf57A%3D%3D"
my_key_decode = requests.utils.unquote(my_key)
base_url = "http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcNrgTrade?"

LAWD_CD = "11110"
DEAL_YMD = "201512"

req_params = {"serviceKey":my_key_decode,
              "LAWD_CD":LAWD_CD,
              "DEAL_YMD":DEAL_YMD}
r = requests.get(base_url, params = req_params)
xml_str = r.text

# xml -> dict -> json
# json.dumps(xmltodict.parse(xml_str)) # 한글 깨짐
json_str = json.dumps(xmltodict.parse(xml_str), ensure_ascii=False) # ensure_ascii=False 옵션 추가하여 한글 보이게 함
json_object = json.loads(json_str) # str -> json

item = json_object["response"]["body"]["items"]["item"]
df = pd.DataFrame(item)
df
```
