### 1. 데이터 수집 준비

```python
import requests
from bs4 import BeautifulSoup

headers = {"user-agent":"Mozilla/5.0"} # 네이버는 headers가 반드시 필요!
```

### 2. 웹에서 html 가져오기

```python
url = "https://finance.naver.com/item/main.nhn?code=005930"
html = requests.get(url, headers=headers)
html.text
```

html.parser 적용

```python
# html 형식으로 가져옴
soup = BeautifulSoup(html.text)
soup
```

### 3. html에서 원하는 데이터 추출하기
1. find
2. select

```python
sub_section = soup.find_all("tbody") # tbody가 리스트 형식으로 sub_section에 담긴다
#len(sub_section)
tbody = sub_section[2]
td = tbody.find_all("td")
data = [item.get_text().strip() for item in td] # text만 뽑아서 리스트에 추가
```

```python
data
```

```python
# 10개씩 잘라서 result_list에 추가
#n = 10

# 1. for문 돌려서 10개씩 자르기
# result_list = []
#for row in range((len(data) + n - 1) // n):
#    result_list.append(data[row*n:(row+1)*n])

# 2. 첫번째와 비슷한 방법
#for i in range(0, len(data), 10): 0부터 리스트 길이까지 10번씩 반복
#    price.append(tmp[i:i+10])

# 3. reshape으로 나눠보기
import numpy as np    

result_list = np.reshape(data, (-1,10))
result_list
```

```python
### 5. 데이터프레임 만들기
```

```python
import pandas as pd
df = pd.DataFrame(result_list)
df
```

### 6. index 만들기

```python
# row 가져오기
th = tbody.find_all("th", {"scope":"row"})
row_list = [row.text for row in th]

# 데이터프레임에 추가
df["구분"] = row_list
```

```python
df.set_index("구분", inplace=True) # inplace=True를 설정하지 않으면 원본 df는 바뀌지 않음
# df = df.set_index 로도 원본 df 변경 가능
df
```

### 7. columns 이름 만들기

```python
thead = soup.find_all("thead")[2]
scope_col = thead.find_all("th", {"scope":"col"})
col_list = [col.text.strip() for col in scope_col]
col_list = col_list[2:12] # 날짜만 뽑아오기

# 데이터 프레임에 열 이름으로 추가
df.columns = col_list
df
```
