```python
# 정리해서 하나의 for문으로 데이터를 가져와서 dataFrame에 넣기
import os
import pandas as pd
import requests
from bs4 import BeautifulSoup
from PIL import Image

headers = { "User-Agent":"Mozilla/5.0" }
url = "http://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001"
html = requests.get(url, headers = headers).text
soup = BeautifulSoup(html, "html.parser")
all = soup.find("ul", {"class": "type06_headline"})
dl = all.find_all("dl")
news_img = []
news_link = []
news_title = []
news_content = []

for item2 in dl:
    try:
        # 이미지
        img = item2.find("dt", {"class": "photo"}).find("img")
        news_img.append(img["src"])
        
        # 뉴스 링크
        link = item2.find("dt", {"class": ""}).find("a")
        news_link.append(link["href"])
        
        ## 뉴스 제목
        title = link.text.replace("\t", "").replace("\n", "").replace("\r", "")
        news_title.append(title)
        
        # 뉴스 내용
        dd =item2.find("dd")
        content = dd.text.replace("\t", "").replace("\t","").replace("\n", "").split("…")[0]
        
        # 리스트에 추가
        news_content.append(content)
        
    except:
        print("No image")
        

# 폴더 생성
if not os.path.isdir('./img_down'):
    os.mkdir('./img_down')

# 이미지 저장
num = 1
for i in news_img:
    r = requests.get(i)
    print(i)
    file = open('./img_down/photo'+str(num)+'.jpg', "wb")
    file.write(r.content)
    file.close()
    num += 1

# 가져온 데이터 DataFrame에 저장
df = pd.DataFrame(
    {"이미지":news_img,
     "기사 링크":news_link,
     "기사 제목":news_title,
     "기사 내용":news_content})
df[['이미지', '기사 링크', '기사 제목', '기사 내용']]
```

```python
# 다른 방법으로 dataframe 만들기
import os
import pandas as pd
import requests
from bs4 import BeautifulSoup
from PIL import Image

headers = { "User-Agent":"Mozilla/5.0" }
url = "http://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001"
html = requests.get(url, headers = headers).text
soup = BeautifulSoup(html, "html.parser")
all = soup.find("ul", {"class": "type06_headline"})
dl = all.find_all("dl")
news_img = []
news_link = []
news_title = []
news_content = []


# 하나의 리스트에 이미지 링크, 뉴스링크, 뉴스제목, 뉴스내용 다 담기
news = []
for item2 in dl:
    try:
        ## 이미지
        img = item2.find("dt", {"class": "photo"}).find("img")
        
        ## 뉴스 링크
        link = item2.find("dt", {"class": ""}).find("a")
        
        ## 뉴스 제목
        title = link.text.replace("\t", "").replace("\n", "").replace("\r", "")
        
        # 뉴스 내용
        dd =item2.find("dd")
        content = dd.text.replace("\t", "").replace("\t","").replace("\n", "").split("…")[0]
        
        news.append([img["src"], link["href"], title, content])
    except:
        print("No image")


# 이미지를 다운받을 폴더 생성
if not os.path.isdir('./img_down'):
    os.mkdir('./img_down')

# 이미지 폴더에 저장
num = 1
for i in news_img:
    r = requests.get(i)
    print("hi")
    file = open('./img_down/photo'+str(num)+'.jpg', "wb")
    file.write(r.content)
    file.close()
    num += 1

# 가져온 데이터 DataFrame에 저장
df = pd.DataFrame(news, columns=["이미지", "기사 링크", "기사 제목", "기사 내용"])
```
