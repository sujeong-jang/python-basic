```python
import requests
from bs4 import BeautifulSoup
```

```python
# 웹에서 크롤링
url = "https://finance.naver.com/item/main.nhn?code=005930"
response = requests.get(url)
if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

else : 
    print(response.status_code)
    #content > div.section.cop_analysis > div.sub_section > table > thead > tr:nth-child(2) > th:nth-child(1)
```

```python
title = []
price = []
date = []
th = soup.find("table", attrs={"class":"tb_type1 tb_num tb_type1_ifrs"}).select_one("tr:nth-of-type(2)").find_all("th")
tbody = soup.find("table", attrs={"class":"tb_type1 tb_num tb_type1_ifrs"}).find("tbody")
td = tbody.find_all("td")

# 날짜 리스트에 넣기
for d in th:
    tmp = d.text.replace("\n", "").replace("\t","")
    date.append(tmp)

print(date)

```

```python
# 타이틀(매출액, 영업이익, 당기순이익 ...) 리스트에 넣기
for t in tbody:
    if t.find("th") != -1:
        title.append(t.find("th").get_text())
print(title)    
```

```python
# 가격 리스트에 넣기
tmp = []
for item in td:
    tmp.append(item.text.replace("\n", "").replace("\t",""))
```
