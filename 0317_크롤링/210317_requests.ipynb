### request: 파이썬용 HTTP 라이브러리
API
- GET
- POST
- DELETE
- PUT
- HEAD: 서버 리소스의 헤더 (메타데이터 획득)
- OPTIONS
- PATCH

```python
# https://requests.readthedocs.io/en/latest/user/quickstart/
```

```python
import requests # 모듈 가져오기
```

```python
# github api에 get 요청 보내기
r = requests.get('https://api.github.com/events') # request 객체를 얻음
dir(r)
```

```python
r = requests.post('https://httpbin.org/post', data = {'key':'value'}) # 데이터 추가
r = requests.put('https://httpbin.org/put', data = {'key':'value'}) # 데이터 수정
r = requests.delete('https://httpbin.org/delete') # 데이터 삭제
r = requests.head('https://httpbin.org/get')
r = requests.options('https://httpbin.org/get')
```

```python
# url에 매개변수 전달
payload = {'key1': 'value1', 'key2': 'value2'}
r = requests.get('https://httpbin.org/get', params=payload)
```

```python
r
```

```python
r.url
```

```python
r.text
```

```python
# 인코딩 확인
r = requests.get('https://api.github.com/events') # request 객체를 얻음
r.encoding
```

```python
headers = {"User-Agent":"Mozillz/5.0"}
r = requests.get('https://www.naver.com', headers=headers)
r.encoding
```

```python
# 인코딩 변경
r.encoding = 'ISO-8859-1'
```

```python
r.content
```

```python
r = requests.get('https://httpbin.org/get')
r.status_code
```

```python
# 웹 사이트가 정상적으로 동작하는지 확인할때 사용
r.status_code == requests.codes.ok
```
