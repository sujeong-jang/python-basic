### 네이버 증권 시세 DataFrame 만들기
- https://finance.naver.com/item/sise_day.nhn?code=005930&page=1
- 종목 10개, 5 page
- 종목코드, 날짜, 종가, 전일대비, 시가, 고가, 저가, 거래량



```python
import pandas as pd
import requests
import time

#pd.set_option('display.max_rows', None)
headers = {"user-agent":"Mozilla/5.0"}

# 종목코드
code_list = ['005930', '000660','019170', '015760', '011200', '185750', '068270', '035720', '214390', '068290']
df = pd.DataFrame()

for code in code_list:
    for num in range(1,6):
        url = "https://finance.naver.com/item/sise_day.nhn?code=" + code + "&page=" + str(num)
        res = requests.get(url, headers = headers)
        table = pd.read_html(res.text)
        new_table = table[0].dropna(0)
        new_table.insert(0, "종목코드", code)
        df = df.append(new_table, ignore_index=True)
```
