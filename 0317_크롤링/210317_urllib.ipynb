### urllib: URL 처리 모듈
1. urllib.request : URL 요청을 위한 클래스나 함수
2. urllib.parse : URL의 구문을 분석하기 위한 함수
3. urllib.error : urllib.request 모듈에 의해 발생된 예외 처리를 위한 클래스
4. urllib.robotparse : 웹사이트의 robots.txt 파일에 대한 분석

```python
import urllib.request
# with: 자원 f를 획득하고 사용 후 자동으로 반납한다.
with urllib.request.urlopen('http://www.python.org') as f:
    print(f.read(300))
```

```python
import urllib.request
# 바이너리 코드로 가져오기 때문에 utf-8로 decoding 해줘야 한다
with urllib.request.urlopen('http://www.python.org') as f:
    print(f.read(300).decode('utf-8'))
```

```python
import urllib.request
import urllib.parse
params = urllib.parse.urlencode({"mode":"LSD",
                                "mid": "sec",
                                "sid1": "001",
                                "date": "20210317",
                                "page": "3"})

url = "https://news.naver.com/main/list.nhn?%s" % params

# with urllib.request.urlopen(url) as f:
#     print(f.read(300).decode('utf-8'))
    
opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]
res = opener.open(url)
res.read().decode("euc-kr") 
```
