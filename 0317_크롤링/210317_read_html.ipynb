### pd.read_html()
pd.read_html()을 이용하면 html에 있는 table 속성에 해당하는 값을 가져올 수 있다

```python
import pandas as pd
import requests
```

```python
url = "https://finance.naver.com/item/main.nhn?code=005930"
table = pd.read_html(url, encoding='euc-kr')
table # 리스트 안에 dataframe이 들어있는 형태로 모든 table을 한번에 가져온다
```
