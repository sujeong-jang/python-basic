### group by
#### : 데이터를 구분할 수 있는 column값들을 이용하여 데이터를 여러 기준에 의해 구분 한 뒤 계산할 수 있는 방법

```python
german_sample2 = german[['Credit Amount', 'Type of apartment', 'Purpose']]
german_sample2.head(3)
```

```python
german_grouped = german_sample2["Credit Amount"].groupby(german_sample2["Type of apartment"])
german_grouped.mean()
```

```python
german_grouped2 = german_sample2["Credit Amount"].groupby([german_sample2["Purpose"],
                                                          german_sample2["Type of apartment"]])
german_grouped2.min()
```

```python
german_sample3 = german[['Credit Amount', 'Type of apartment', 'Sex & Marital Status']]
german_sample3.head(3)
```

```python
german_sample3.groupby("Type of apartment").mean()
```

```python
# 그룹 간 반복하기
group_data = german_sample3.groupby("Type of apartment")
for type, group in group_data:
    print("type: ", type)
    print("group: ", group.head(3))
```

```python
# 두 개 그룹 반복
group_data = german_sample3.groupby(["Type of apartment", "Sex & Marital Status"])
for (type, sex), group in group_data:
    print((type, sex))
    print(group.head(3))
```
